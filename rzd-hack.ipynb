```python
def predict_bboxes_one_smaple(model, path, ax=None):
    if ax is None:
        plt.imshow(model(path)[0].plot())
    else:
        ax.imshow(model(path)[0].plot())
```

```python
def predict_trained_on_cigarette(model, path, ax=None):
    threshold = 0.4
    cigarette_class = 0
    person_class = 1
    res = model(path)
    img = Image.open(path)
    
    if ax is None:
        plt.imshow(img)
        if (cigarette_class in res[0].boxes.cls.cpu().numpy()) and (person_class in res[0].boxes.cls.cpu().numpy()):
            conf = res[0].boxes.conf.cpu().numpy()[cigarette_class == res[0].boxes.cls.cpu().numpy()]
            conf = max(conf)
            plt.title('Smoker confidence: {:.0f}%'.format(conf * 100))
            # if conf > threshold:
            #     plt.title('Smoker confidence: {:.0f}%'.format(conf * 100))
            # else:
            #     plt.title('Not smoker')
        else:
            plt.title('Not smoker')
    else:
        ax.imshow(img)
        if (cigarette_class in res[0].boxes.cls.cpu().numpy()) and (person_class in res[0].boxes.cls.cpu().numpy()):
            conf = res[0].boxes.conf.cpu().numpy()[cigarette_class == res[0].boxes.cls.cpu().numpy()]
            conf = max(conf)
            ax.set_title('Smoker confidence: {:.0f}%'.format(conf * 100))
            # if conf > threshold:
            #     ax.set_title('Smoker confidence: {:.0f}%'.format(conf * 100))
            # else:
            #     ax.set_title('Not smoker')
        else:
            ax.set_title('Not smoker')
```

```python
def predict_bboxes_some(model, path_to_dir, max_samples=3):
    fig, axes = plt.subplots(1, max_samples, figsize=(15, 7))
    paths = glob.glob(path_to_dir + '/*')
    np.random.shuffle(paths)
    for i, path in enumerate(paths):
        predict_bboxes_one_smaple(model, path, axes[i])
        if i + 1 >= max_samples:
            break
```

```python
def predict_trained_on_cigarette_some(model, path_to_dir, max_samples=3):
    fig, axes = plt.subplots(1, max_samples, figsize=(15, 7))
    paths = glob.glob(path_to_dir + '/*')
    np.random.shuffle(paths)
    for i, path in enumerate(paths):
        predict_trained_on_cigarette(model, path, axes[i])
        if i + 1 >= max_samples:
            break
```
